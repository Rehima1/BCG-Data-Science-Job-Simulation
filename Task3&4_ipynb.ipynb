```python
import pandas as pd

df = pd.read_csv('clean_data_after_eda.csv')
```

```python
df.head()
```

```python
df.drop(columns=['id'], inplace=True)
```

```python
for col in df.columns:
    if df[col].nunique() == 1:
        df.drop(columns=[col], inplace=True)
```

```python
df.head()
```

```python
date_cols = ['date_activ','date_end','date_modif_prod','date_renewal']

for col in date_cols:
  df[col] = pd.to_datetime(df[col], errors ='coerce')


for col in date_cols:
  df[f'{col} year'] = df[col].dt.year
  df[f'{col} month'] = df[col].dt.month
  df[f'{col} day'] = df[col].dt.day

df['activity_duration'] = (df['date_end']- df['date_activ']).dt.days
print(df['activity_duration'])
```

```python
df.drop(columns=date_cols, inplace=True)

# Check
print(df.dtypes)
```

```python
df.isnull().sum()
```

```python
skewed = [
    'cons_12m',
    'cons_gas_12m',
    'cons_last_month',
    'forecast_cons_12m',
    'forecast_cons_year',
    'forecast_discount_energy',
    'forecast_meter_rent_12m',
    'forecast_price_energy_off_peak',
    'forecast_price_energy_peak',
    'forecast_price_pow_off_peak'
]

df[skewed].describe()
```

```python
import numpy as np

df["cons_12m"] = np.log10(df["cons_12m"] + 1)
df["cons_gas_12m"] = np.log10(df["cons_gas_12m"] + 1)
df["cons_last_month"] = np.log10(df["cons_last_month"] + 1)
df["forecast_cons_12m"] = np.log10(df["forecast_cons_12m"] + 1)
df["forecast_cons_year"] = np.log10(df["forecast_cons_year"] + 1)
df["forecast_meter_rent_12m"] = np.log10(df["forecast_meter_rent_12m"] + 1)
df["imp_cons"] = np.log10(df["imp_cons"] + 1)
```

```python
df[skewed].describe()
```

```python
corr=df.corr()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(45, 45))
sns.heatmap(corr, xticklabels=corr.columns.values,
    yticklabels=corr.columns.values, annot = True)
plt.xticks(fontsize=18)
plt.yticks(fontsize=18)
plt.show()
```

```python
df = pd.get_dummies(df , columns=['channel_sales', 'origin_up', 'has_gas'], drop_first= True)
```

Model Building

```python
from sklearn.model_selection import train_test_split

x = df.drop(columns =['churn'])
y = df['churn']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2 , random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report

model = RandomForestClassifier(random_state=42)

model.fit(x_train, y_train)

y_prediction = model.predict(x_test)

accuracy = accuracy_score(y_test, y_prediction)
print(f'the accuracy is {accuracy}')
print(classification_report(y_test, y_prediction))
```
